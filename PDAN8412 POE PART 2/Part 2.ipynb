# Theory

## A. Explanation of what Logistic Regression is.


## What is Logistic Regression from scratch:
### Logistic regression is a supervised learning approach for binary classification that predicts whether a target variable belongs to one of two categories. It employs the logistic (sigmoid) function to convert input data into probabilities ranging from 0 to 1. The method weights and biases the input characteristics, then linearly combines them before passing the result via the sigmoid function, which returns a probability score. This probability is compared to a threshold (usually 0.5) to categorize the input. Logistic regression iteratively updates the weights using optimization (for example, gradient descent) to reduce prediction error and refine the model's decision boundary.(geeksforgeeks, 2024)

## How does Logistic Regression work?
### The logistic regression model uses a sigmoid function to convert the continuous value output of the linear regression function to categorical value output. This function transfers any real-valued collection of independent variables input into a value between 0 and 1. This function is referred to as the logistic function.(geeksforgeeks, 2024)

## Avantages of logistic Regression:
### 1. Ease of Implementation and Interpretation:
 - #### Logistic regression is simple to apply and interpret, making it an accessible alternative for novices. It is also incredibly efficient in terms of training time, making it ideal for datasets that require speed.(geeksforgeeks, 2024)

### 2. No Assumptions About Class Distribution:
 - #### Not like other models, logistic regression makes no assumptions about class distributions in the feature space. This versatility enables it to handle a wide range of datasets.

### 3. Extension to Multinomial Cases:
 - #### Logistic regression may be expanded to cover many classes using multinomial regression. It provides a natural probabilistic framework for class prediction, providing for a better understanding of the model's judgments.(geeksforgeeks, 2024)

### 4. Coefficient Interpretation: 
 - #### Logistic regression not only provides a measure of a predictor's suitability (by coefficient size), but it also reveals the direction of the association (positive or negative), making the findings more interpretable.(geeksforgeeks, 2024)

## Disadvanatge of Logistic Regression:
### 1. Risk of Overfitting:
 - #### When the number of observations is less than the number of features, logistic regression may overfit. This makes it less suited for high-dimensional datasets, unless regularization is used.(geeksforgeeks, 2024)

### 2. Linear Boundaries:
 - #### Logistic regression creates linear boundaries, making it unsuitable for datasets with complicated or nonlinear relationships between variables.(geeksforgeeks, 2024)

### 3. Assumption of Linearity:
 - #### One fundamental disadvantage of logistic regression is that it presupposes a linear connection between the independent variables and the log chances of the outcome. This can diminish its efficacy in circumstances when the link is not linear.(geeksforgeeks, 2024)

### 4. Limited to Discrete Outcomes:
 - #### Logistic regression can only predict discrete outcomes, which means that the dependent variable must be categorical. It cannot be utilized with continuous outcome variables, which limits its use in some situations.(geeksforgeeks, 2024)

## Logistic Regression from scratch

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay,f1_score, classification_report
```

### Creating the LogisticRegression from Scratch code.

```python
# ***************************************************************************************
#    Title: Implementing logistic regression from scratch in Python
#    Author: Casper Hansen
#    Date Published: 14 February 2022
#    Date Retrieved: 12 October 2024
#    Code version: 1
#    Availability: https://www.geeksforgeeks.org/create-a-correlation-matrix-using-python/
#
# ***************************************************************************************


# The 'sigmoid' function is used in logistic regression to introduce non-linearity into the model.
# It maps any real-valued number into a probability between 0 and 1.
# The 'sigmoid' function with input values clipped to prevent overflow.
def sigmoid(x):
    # Clip the input values to prevent overflow in the exponential function.
    # Clipping to a reasonable range like -500 to 500 helps manage very large values.
    x = np.clip(x, -500, 500)
    return 1 / (1 + np.exp(-x))

# Define the Logistic Regression class with gradient descent.
class LogisticRegressionScratch:
    def __init__(self, lr=0.001, n_iters=1000):
        self.lr = lr
        self.n_iters = n_iters
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Gradient Descent for training the model.
        for _ in range(self.n_iters):
            linear_pred = np.dot(X, self.weights) + self.bias
            predictions = sigmoid(linear_pred)
            
            # Calculate gradients
            dw = (1 / n_samples) * np.dot(X.T, (predictions - y))
            db = (1 / n_samples) * np.sum(predictions - y)
            
            # Update weights and bias
            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        linear_pred = np.dot(X, self.weights) + self.bias
        y_pred = sigmoid(linear_pred)
        class_pred = [0 if y <= 0.5 else 1 for y in y_pred]
        return class_pred
```

### Setting x and y and doing training and spliting and testing the model.

```python

# SETTING UP x AND y
x = df_pandas_selected.drop(['pha'], axis=1) # LET x BE DATA WITHOUT 'pha'
y = df_pandas_selected['pha'] # LET y BE 'pha

# THIS SPLITS THE DATA, test_size=0.3 OR 30%, AND THE TRAIN SIZE IS EQUAL TO THE REMAINING 70%
# WHY random_state=42, IS TO ENSURE THAT THE OUTPUT THAT IS RUN EACH TIME EQUALS TO THE FIRST RUN OUTPUT WHEN SPLITTING

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, shuffle=True, stratify=y, random_state=42)
print(x_train.shape) # SHOWS THE SIZE OF x_train
print(x_test.shape) # SHOWS THE SIZE OF x_test
```

### Setting up the Logistic Regression model from sratch.

```python

# Create the Logistic Regression model object
classifier = LogisticRegressionScratch(lr=0.001) # With learning rate of 0.01

# Fit the model with training data
classifier.fit(x_train, y_train)

# Generate predictions on the test set
predictions = classifier.predict(x_test)
```

### Method for calculating accuracy of the model , f1 score and displaying classification report

```python
# Method for calculating accuracy of the model
def accuracy(predictions, y_test): 
    return np.sum(predictions == y_test) / len(y_test)

# Calculate and display accuracy
acc = accuracy(predictions, y_test)
print(f"Accuracy: {acc * 100:.2f}%")

# Calculate F1 score with 
f1 = f1_score(y_test, predictions, average="weighted")

# Display F1 score
print(f"\nF1 Score: {f1 * 100:.2f}%")

# Generate classification report 
classification_rep = classification_report(y_test, predictions)

# Display the classification report
print("\nClassification Report:\n", classification_rep)
```

Precision is the ratio of successfully predicted positive observations to total predicted positives for each class, whereas recall (or sensitivity) is the ratio of correctly predicted positive observations to real positives. The F1 score, which is the harmonic mean of accuracy and recall, balances their contributions on a scale of 0 to 1, with 1 being optimal. Support is the number of instances of each class in the dataset.

Despite its great accuracy, the model's failure to accurately categorize any occurrences of class 1 significantly restricts its utility. The confusion matrix underlines this issue by concentrating predictions in the True Negative and False Negative categories, implying that the model is overfitting to the dominant class (class 0) while failing to generalize to the minority class (class 1). To address this issue, approaches such as oversampling or undersampling may be required, as well as the use of alternative algorithms capable of dealing with class imbalance more successfully.

```python
# ***************************************************************************************
#    Title: Handling Imbalanced Data for Classification
#    Author: geeksforgeeks
#    Date Published: 02 January 2024
#    Date Retrieved: 12 October 2024
#    Code version: 1
#    Availability: https://www.geeksforgeeks.org/create-a-correlation-matrix-using-python/
#
# ***************************************************************************************

# Random Undersampling
# Combine X and y for easier manipulation
train_data = pd.concat([x_train, y_train], axis=1)

# Separate majority and minority classes
majority = train_data[train_data['pha'] == 0]
minority = train_data[train_data['pha'] == 1]

# Randomly downsample majority class
majority_downsampled = resample(majority,
                                 replace=False,     # sample without replacement
                                 n_samples=len(minority),  # to match minority class
                                 random_state=42)  # reproducible results

# Combine minority class with downsampled majority class
downsampled = pd.concat([majority_downsampled, minority])
```

```python
# Split back into X and y
x_train_balanced = downsampled.drop(['pha'], axis=1)
y_train_balanced = downsampled['pha']
```

### Create a pipeline to handle StandardScaler and LogisticRegression

```python
# Create a Logistic Regression model with class weights
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scale features for better performance
    ('log_reg', LogisticRegression(class_weight='balanced'))  # Logistic Regression with class weights
])
```

### Setting hyperparameters for GridSearchCV
 

### why use GridSearchCV?
#### The provided code defines hyperparameters for GridSearchCV, which is used for hyperparameter tuning in logistic regression. The param_grid dictionary specifies a range of values for the regularization strength (log_reg__C), the regularization technique (log_reg__penalty), and the solvers to be tested (log_reg__solver). This setup allows for systematic exploration of combinations to optimize the model's performance.

```python
# Define hyperparameters for GridSearchCV
param_grid = {
    'log_reg__C': [0.01, 0.1, 1, 10, 100],  # Regularization strength
    'log_reg__penalty': ['l2'],  # Regularization technique (l2 is default)
    'log_reg__solver': ['liblinear', 'saga']  # Solvers for logistic regression
}
```

## Performing the GridSearchCV 

```python
# Perform GridSearchCV to find the best parameters
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='f1', n_jobs=-1, verbose=2)
grid_search.fit(x_train_balanced, y_train_balanced)

# Get the best model from the grid search
best_model = grid_search.best_estimator_

# Make predictions with the best model
predictions = best_model.predict(x_test)

# Evaluate model performance
accuracy = accuracy_score(y_test, predictions)
f1 = f1_score(y_test, predictions)
class_report = classification_report(y_test, predictions)
```
